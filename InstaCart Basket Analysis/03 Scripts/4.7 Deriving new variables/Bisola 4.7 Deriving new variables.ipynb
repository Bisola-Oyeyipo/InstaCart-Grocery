### 04. User-defined function

```python
# defining a function with a if statement
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

### 05. if statement with loc() function

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

### 08. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day.

```python
# viewing the column in the dataframe
ords_prods_merge.head()
```

```python
ords_prods_merge.tail()
```

```python
# Check value counts in "order_hour_of_day" column.
ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
# Create if statement with for loops labeling the hour of the day
#the hour of the day will be subdivided into three which will be most order, fewest orders, average.
busiest_hour = []
for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]: busiest_hour.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]: busiest_hour.append("Fewest orders")
  else: busiest_hour.append("Average orders")
```

```python
busiest_hour
```

```python
# Create new column "busiest_period_of_day" in ords_prods_merge.
ords_prods_merge['busiest_period_of_day'] = busiest_hour
```

#### 05. Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day

```python
#create a sample of the dataframe
sample_df = df.sample(n = 100000, random_state = 1)
```

```python
#group by hour_of_the_day and calculate the mean expenditure
hour_expenditure = sample_df.groupby('order_hour_of_day')['prices'].mean().reset_index()
```

```python
# create a line chart
line_price_hod = sns.lineplot(data = sample_df, x = 'order_hour_of_day', y = 'prices')
```

#### 06. Determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents

```python
sample_df.columns
```

```python
# create a line chart
#nod means the number of dependent
line_age_nod = sns.lineplot(data = sample_df, x = 'age', y = 'number_of_dependents')
```

There is no clear correlation between the age and number of dependents.

#### 07. Explore whether there’s a connection between age and spending power (income)

```python
# create a scatterplot of 'Age' and 'Income'
scatter = sns.scatterplot(x='age', y='income', data = sample_df)
```

There is a higher concentration on the income below 200k for all age group and there was a slight increase in the income from age 40. From the plot, it shows that the younger generation(between 0 - 40) have a lesser  income in comparison to age 40 to 80 that can earn up to 600000 as income.

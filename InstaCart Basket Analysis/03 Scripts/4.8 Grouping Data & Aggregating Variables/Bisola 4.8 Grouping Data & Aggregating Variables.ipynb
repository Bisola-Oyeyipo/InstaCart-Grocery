### 04. creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Creating new column "max_order"
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking the Output
ords_prods_merge.head(20)
```

```python
# Creating a 'Loyalty' Label Flag using loc Function
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Examining Value Counts on 'loyalty_flag'
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking if loyalty flag have been successfully created
ords_prods_merge.head(10)
```

### Question 5:Using the loyalty flag, check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer)

```python
# checking the basic statistics of the product prices for each loyalty category
ords_prods_merge.groupby(['loyalty_flag']).agg({'prices':['mean','min','max']})
```

Loyql customers have the least price of products while New customer have the highest price of products but just a little higher price than a regular customer.

### Question 6: Create a spending flag for each user based on the average price across all their orders.

```python
# Create a column for average spending customer
ords_prods_merge['average_spend']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create a flag for low and high spender
ords_prods_merge.loc[ords_prods_merge['average_spend'] < 10, 'spender_flag'] = 'Low_spender'
ords_prods_merge.loc[ords_prods_merge['average_spend'] >= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check the value count for spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
#checking the output of the average spend_
ords_prods_merge[['user_id', 'average_spend', 'spender_flag', 'prices']].head(20)
```

### Question 7: Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column

```python
# Create a column for frequent customers
ords_prods_merge['customer_regularity'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking the output for the new column 
ords_prods_merge[['user_id', 'days_since_prior_order', 'customer_regularity']].head(10)
```

```python
# Create a flag for order frequency
ords_prods_merge.loc[ords_prods_merge['customer_regularity'] > 20, 'frequency_flag'] = 'Non_frequent_customer'
ords_prods_merge.loc[(ords_prods_merge['customer_regularity'] > 10) & (ords_prods_merge['customer_regularity'] <= 20),'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['customer_regularity'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check the value count of frequency_flag
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
#checking some relevant columns
ords_prods_merge[['user_id','days_since_prior_order', 'customer_regularity', 'frequency_flag']].head(100)
```
